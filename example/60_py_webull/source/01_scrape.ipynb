```python
# open browser that can be automated with python
# zoom out, so that everything is rendered
fp = webdriver.FirefoxProfile()
fp.set_preference("layout.css.devPixelsPerPx", ".5")
driver = webdriver.Firefox(fp)
driver.set_window_size(((2) * (1900)), ((2) * (1040)))
driver.get("https://app.webull.com/market/region/6")
```

```python
# wait up to 3 seconds for elements to appear
driver.implicitly_wait(3)
```

```python
a = driver.find_element_by_xpath(
    "/html/body/div/div/div[3]/div/div[1]/div/div[2]/div/div/div[3]/div[1]/span/div/div"
)
a.click()
```

```python
# get header title with sector name
header = driver.find_element_by_xpath(
    "/html/body/div/div/div[3]/div/div[1]/div/div[2]/div[1]/div[2]/div/div/div[1]/div[3]/div[1]/div[1]/span[1]"
)
perc = driver.find_element_by_xpath(
    "/html/body/div/div/div[3]/div/div[1]/div/div[2]/div[1]/div[2]/div/div/div[1]/div[3]/div[1]/div[1]/span[2]"
)
# get growth percentage for the entire sector
display(header.text)
display(perc.text)
```

```python
# get all sector names from the grid of tables
res = []
for col in range(1, ((1) + (3))):
    for row in range(1, ((1) + (5))):
        header = driver.find_element_by_xpath(
            "/html/body/div/div/div[3]/div/div[1]/div/div[2]/div[1]/div[2]/div/div/div[1]/div[{}]/div[{}]/div[1]/span[1]"
            .format(row, col))
        perc = driver.find_element_by_xpath(
            "/html/body/div/div/div[3]/div/div[1]/div/div[2]/div[1]/div[2]/div/div/div[1]/div[{}]/div[{}]/div[1]/span[2]"
            .format(row, col))
        res.append(
            dict(row=row, col=col, sector=header.text,
                 sector_growth=perc.text))
df = pd.DataFrame(res)
df.sort_values(by="sector_growth", ascending=False)
```
